```python
# creating Webcrawler
useragents={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'}
dataframelist=[]

Number_of_pages=range(0,14) #Initializing range to publication papers  
for pages in Number_of_pages:
    seed='https://pureportal.coventry.ac.uk/en/organisations/school-of-life-sciences/publications/?page={}'.format(pages)
    # We have used a BFS approach
    # considering the structure as
    # a tree. It uses a queue based
    # approach to traverse
    # links upto a particular depth.
    queue = [seed] 
    count = 0 #initializing the seed to our dataset  
    while(queue!=[] and count < 14):
        count +=1
        url = queue.pop(0)
        print(seed)
    x = requests.get(url,headers=useragents)
    y = x.text
    soup = BeautifulSoup(y, "html.parser")
    publications = soup.find_all('div',{'class':'result-container'})
    for all_pub in publications:
        try:
            result={
                'title':all_pub.find('a',{'class':'link'}).text,
                'link':all_pub.find('a',{'class':'link'})['href'],
                'author':all_pub.find('a',{'class':'link person'}).text,
                'author_link':all_pub.find('a',{'class':'link person'})['href'],
                'date':all_pub.find('span',{'class':'date'}).text
                }
           
        except AttributeError:
             result={
                'title':all_pub.find('a',{'class':'link'}).text,
                'link':all_pub.find('a',{'class':'link'})['href'],
                'author':None,
                'author_link':None,
                'date':all_pub.find('span',{'class':'date'}).text
                }
        dataframelist.append(result)
        dataframe = pd.DataFrame(dataframelist)
        dataframe.to_csv('slspublications.csv')
        #print(result) #uncomment this to show total 14 pages publications of sls members
```

```python
import nltk
from nltk.corpus import stopwords
import string
import re

cleaning_doc = []
for frame in SLS_data['title']:
   
    doc = re.sub(r'[^\x00-\x7F]+', '', frame)  # Unicode removal
    
    doc = re.sub(r'@\w+', '', doc) # Mentions removal
   
    doc = doc.lower()  # convert the document into lowercase
 
    doc = re.sub(r'[%s]' % re.escape(string.punctuation), '', doc)  # punctuations removal
 
    doc = re.sub(r'[0-9]', '', doc)  # convert the numbers into lowercase
   
    doc = re.sub(r'\s{2,}', '', doc)  # double space removal

    pattern = re.compile(r'\b(' + r'|'.join(stopwords.words('english')) + r')\b\s*')     # stop words removal
    text = pattern.sub('', doc)
    cleaning_doc.append(text)
```

```python
from functools import reduce
_WORD_MIN_LENGTH = 3
_STOP_WORDS = frozenset([
'a', 'about', 'above', 'above', 'across', 'after', 'afterwards', 'again', 
'against', 'all', 'almost', 'alone', 'along', 'already', 'also','although',
'always','am','among', 'amongst', 'amoungst', 'amount',  'an', 'and', 'another',
'any','anyhow','anyone','anything','anyway', 'anywhere', 'are', 'around', 'as',
'at', 'back','be','became', 'because','become','becomes', 'becoming', 'been', 
'before', 'beforehand', 'behind', 'being', 'below', 'beside', 'besides', 
'between', 'beyond', 'bill', 'both', 'bottom','but', 'by', 'call', 'can', 
'cannot', 'cant', 'co', 'con', 'could', 'couldnt', 'cry', 'de', 'describe', 
'detail', 'do', 'done', 'down', 'due', 'during', 'each', 'eg', 'eight', 
'either', 'eleven','else', 'elsewhere', 'empty', 'enough', 'etc', 'even', 
'ever', 'every', 'everyone', 'everything', 'everywhere', 'except', 'few', 
'fifteen', 'fify', 'fill', 'find', 'fire', 'first', 'five', 'for', 'former', 
'formerly', 'forty', 'found', 'four', 'from', 'front', 'full', 'further', 'get',
'give', 'go', 'had', 'has', 'hasnt', 'have', 'he', 'hence', 'her', 'here', 
'hereafter', 'hereby', 'herein', 'hereupon', 'hers', 'herself', 'him', 
'himself', 'his', 'how', 'however', 'hundred', 'ie', 'if', 'in', 'inc', 
'indeed', 'interest', 'into', 'is', 'it', 'its', 'itself', 'keep', 'last', 
'latter', 'latterly', 'least', 'less', 'ltd', 'made', 'many', 'may', 'me', 
'meanwhile', 'might', 'mill', 'mine', 'more', 'moreover', 'most', 'mostly', 
'move', 'much', 'must', 'my', 'myself', 'name', 'namely', 'neither', 'never', 
'nevertheless', 'next', 'nine', 'no', 'nobody', 'none', 'noone', 'nor', 'not', 
'nothing', 'now', 'nowhere', 'of', 'off', 'often', 'on', 'once', 'one', 'only',
'onto', 'or', 'other', 'others', 'otherwise', 'our', 'ours', 'ourselves', 'out',
'over', 'own','part', 'per', 'perhaps', 'please', 'put', 'rather', 're', 'same',
'see', 'seem', 'seemed', 'seeming', 'seems', 'serious', 'several', 'she', 
'should', 'show', 'side', 'since', 'sincere', 'six', 'sixty', 'so', 'some', 
'somehow', 'someone', 'something', 'sometime', 'sometimes', 'somewhere', 
'still', 'such', 'system', 'take', 'ten', 'than', 'that', 'the', 'their', 
'them', 'themselves', 'then', 'thence', 'there', 'thereafter', 'thereby', 
'therefore', 'therein', 'thereupon', 'these', 'they', 'thickv', 'thin', 'third',
'this', 'those', 'though', 'three', 'through', 'throughout', 'thru', 'thus', 
'to', 'together', 'too', 'top', 'toward', 'towards', 'twelve', 'twenty', 'two', 
'un', 'under', 'until', 'up', 'upon', 'us', 'very', 'via', 'was', 'we', 'well', 
'were', 'what', 'whatever', 'when', 'whence', 'whenever', 'where', 'whereafter',
'whereas', 'whereby', 'wherein', 'whereupon', 'wherever', 'whether', 'which', 
'while', 'whither', 'who', 'whoever', 'whole', 'whom', 'whose', 'why', 'will', 
'with', 'within', 'without', 'would', 'yet', 'you', 'your', 'yours', 'yourself',
'yourselves', 'the'])

def word_split(text):
    """
    Split a text in words. Returns a list of tuple that contains
    (word, location) location is the starting byte position of the word.
    """
    word_list = []
    wcurrent = []
    windex = None

    for i, c in enumerate(text):
        if c.isalnum():
            wcurrent.append(c)
            windex = i
        elif wcurrent:
            word = u''.join(wcurrent)
            word_list.append((windex - len(word) + 1, word))
            wcurrent = []

    if wcurrent:
        word = u''.join(wcurrent)
        word_list.append((windex - len(word) + 1, word))

    return word_list

def words_cleanup(words):
    """
    Remove words with length less then a minimum and stopwords.
    """
    cleaned_words = []
    for index, word in words:
        if len(word) < _WORD_MIN_LENGTH or word in _STOP_WORDS:
            continue
        cleaned_words.append((index, word))
    return cleaned_words

def words_normalize(words):
    """
    Do a normalization precess on words. In this case is just a tolower(),
    but you can add accents stripping, convert to singular and so on...
    """
    normalized_words = []
    for index, word in words:
        wnormalized = word.lower()
        normalized_words.append((index, wnormalized))
    return normalized_words

def word_index(text):
    """
    Just a helper method to process a text.
    It calls word split, normalize and cleanup.
    """
    words = word_split(text)
    words = words_normalize(words)
    words = words_cleanup(words)
    return words

def inverted_index(text):
    """
    Create an Inverted-Index of the specified text document.
        {word:[locations]}
    """
    inverted = {}

    for index, word in word_index(text):
        locations = inverted.setdefault(word, [])
        locations.append(index)

    return inverted

def inverted_index_add(inverted, doc_id, doc_index):
    """
    Add Invertd-Index doc_index of the document doc_id to the 
    Multi-Document Inverted-Index (inverted), 
    using doc_id as document identifier.
        {word:{doc_id:[locations]}}
    """
    for word, locations in doc_index.items():
        indices = inverted.setdefault(word, {})
        indices[doc_id] = locations
    return inverted

def search(inverted, query):
    """
    Returns a set of documents id that contains all the words in your query.
    """
    words = [word for _, word in word_index(query) if word in inverted]
    results = [set(inverted[word].keys()) for word in words]
    return reduce(lambda x, y: x & y, results) if results else []

if __name__ == '__main__':
    document1 = """
   title                        link                                                                   author          author_link                                              date
   A Comparison of Physical
   Activity Between Home-
   Based and Centre-Based
   Pulmonary Rehabilitation:                                                                                                                                                   26
   A Randomised Controlled https://pureportal.coventry.ac.uk/en/publications/a-comparison-of-                          https://pureportal.coventry.ac.uk/en/persons/elizabeth- Oct
 0Secondary Analysis            physical-activity-between-home-based-and-centre-b                      Horton, E.      horton                                                  2021
   Actual and perceived
   motor competence
   mediate the relationship
   between physical fitness
   and technical skill                                                                                                                                                          11
   performance in young         https://pureportal.coventry.ac.uk/en/publications/actual-and-                          https://pureportal.coventry.ac.uk/en/persons/michael-    Jul
 1soccer players                perceived-motor-competence-mediate-the-relationship-be                 Duncan, M.      duncan                                                   2021
   Acute Hormonal Response
   to Kettlebell Swing
   Exercise Differs
   Depending on Load, Even                                                                                                                                                      1
   When Total Work Is           https://pureportal.coventry.ac.uk/en/publications/acute-hormonal-                      https://pureportal.coventry.ac.uk/en/persons/derek-      Apr
 2Normalized                    response-to-kettlebell-swing-exercise-differs-depe                     Renshaw, D.     renshaw                                                  2021
   Associations Between
   Self-Reported Sleep
   Duration and Mortality in
   Employed Individuals:
   Systematic Review and        https://pureportal.coventry.ac.uk/en/publications/associations-                        https://pureportal.coventry.ac.uk/en/persons/laura-      Jul
 3Meta-Analysis                 between-self-reported-sleep-duration-all-cause-and-c                   Roden, L. C.    roden                                                    2021
   Blood cancer care in a
   resource limited setting
   during the Covid-19                                                                                                                                                            17
   outbreak; a single center https://pureportal.coventry.ac.uk/en/publications/blood-cancer-care-                                                                                 Sep
 4experience from Sri Lanka in-a-resource-limited-setting-during-the-covid-                            Lin, T. T.      https://pureportal.coventry.ac.uk/en/persons/thet-thet-lin 2021
   British Soccer Academy
   Personnel Perceive
   Psychological and
   Technical/Tactical
   Attributes as the Most                                                                                                                                                       28
   Important Contributors to https://pureportal.coventry.ac.uk/en/publications/british-soccer-                         https://pureportal.coventry.ac.uk/en/persons/richard-    Jul
 5Development                   academy-personnel-perceive-psychological-and-techn                     Kite, R. J.     james-kite                                               2021
   Changes in joint
   kinematics and dynamic
   postural stability with free
   and restricted arm           https://pureportal.coventry.ac.uk/en/publications/changes-in-joint-                    https://pureportal.coventry.ac.uk/en/persons/maximilian-Jul
 6movements in children         kinematics-and-dynamic-postural-stability-with-f                       Wdowski, M.     wdowski                                                 2021
   Comparative analysis of
   reported physical activity
   from leisure centres’
   members versus the                                                                                                                                                           29
   general population in        https://pureportal.coventry.ac.uk/en/publications/comparative-         Lopez           https://pureportal.coventry.ac.uk/en/persons/jorge-      Jun
 7Spain                         analysis-of-reported-physical-activity-from-leisure-c                  Fernandez, J.   lopez-fernandez                                          2021
   Cross-cultural comparison
   of fundamental movement
   skills in 9- to 10- year old                                                                                                                                                9
   children from England and https://pureportal.coventry.ac.uk/en/publications/cross-cultural-                                                                                 Oct
 8China                         comparison-of-fundamental-movement-skills-in-9-to-                     Ma, J.          https://pureportal.coventry.ac.uk/en/persons/jiani-ma-2 2021
   Delivery Approaches
   Within Exercise Referral
   Schemes: A Survey of
   Current Practice in          https://pureportal.coventry.ac.uk/en/publications/delivery-approaches-                 https://pureportal.coventry.ac.uk/en/persons/nikita-     Apr
 9England                       within-exercise-referral-schemes-a-survey-of-                          Rowley, N.      rowley                                                   2021
   Developing Powerful
   Athletes Part 2: Practical https://pureportal.coventry.ac.uk/en/publications/developing-powerful-                   https://pureportal.coventry.ac.uk/en/persons/peter-      Feb
10Applications                  athletes-part-2-practical-applications                                 Mundy, P.       mundy                                                    2021
   Does Resisted Sled
   Towing Improve the
   Physical Qualities of Elite
   Youth Soccer Players of https://pureportal.coventry.ac.uk/en/publications/does-resisted-sled-                                                                                Feb
11Differing Maturity Status? towing-improve-the-physical-qualities-of-elite                            Morris, R.      https://pureportal.coventry.ac.uk/en/persons/rhys-morris 2021
   Dose response effects of
   the BWF Shuttle Time
   Programme on children’s
   actual and perceived
   fundamental movement         https://pureportal.coventry.ac.uk/en/publications/dose-response-                       https://pureportal.coventry.ac.uk/en/persons/michael-    Dec
12skill competence              effects-of-the-bwf-shuttle-time-programme-on-childr                    Duncan, M.      duncan                                                   2021
   Effectiveness of Structured
   Physical Activity
   Interventions Through the
   Evaluation of Physical
   Activity Levels, Adoption,
   Retention, Maintenance,
   and Adherence Rates: A
   Systematic Review and
   Meta-Analysis                https://pureportal.coventry.ac.uk/en/publications/effectiveness-of-                    https://pureportal.coventry.ac.uk/en/persons/nadja-      Jan
13                              structured-physical-activity-interventions-throu                       Willinger, N.   willinger                                                2021
   Effect of Barbell Load on                                                                                                                                                    1
   Vertical Jump Landing        https://pureportal.coventry.ac.uk/en/publications/effect-of-barbell-                   https://pureportal.coventry.ac.uk/en/persons/peter-      Jan
14Force-Time Characteristicsload-on-vertical-jump-landing-force-time-charac                            Mundy, P. D.    mundy                                                    2021
   Effect of sex and fatigue
   on quiet standing and
   dynamic balance and                                                                                                                                                         20
   lower extremity muscle       https://pureportal.coventry.ac.uk/en/publications/effect-of-sex-and-                                                                           Oct
15stiffness                     fatigue-on-quiet-standing-and-dynamic-balance-a                        Hill, M.        https://pureportal.coventry.ac.uk/en/persons/mathew-hill2021
   Enhancing the
   implementation and
   sustainability of
   fundamental movement
   skill interventions in the
   UK and Ireland: lessons
   from collective intelligence                                                                                                                                                    3
   engagement with               https://pureportal.coventry.ac.uk/en/publications/enhancing-the-                                                                                  Nov
16stakeholders                   implementation-and-sustainability-of-fundamental-mo                      Ma, J.           https://pureportal.coventry.ac.uk/en/persons/jiani-ma-2 2021
   Fundamental Movement
   Skill Proficiency Among
   British Primary School
   Children: Analysis at a
   Behavioral Component          https://pureportal.coventry.ac.uk/en/publications/fundamental-                                                                                  Apr
17Level                          movement-skill-proficiency-among-british-primary-scho                    Eyre, E. L. J.   https://pureportal.coventry.ac.uk/en/persons/emma-eyre2021
   Fundamental movement
   skills and perceived
   competence, but not
   fitness, are the key factors
   associated with technical
   skill performance in boys                                                                                                                                                       7
   who play grassroots           https://pureportal.coventry.ac.uk/en/publications/fundamental-                            https://pureportal.coventry.ac.uk/en/persons/michael-   Apr
18soccer                         movement-skills-and-perceived-competence-but-not-fitn                    Duncan, M.       duncan                                                  2021
   Habitual Caffeine
   Consumption Does Not
   Affect the Ergogenicity of
   Coffee Ingestion During a https://pureportal.coventry.ac.uk/en/publications/habitual-caffeine-                                                                                   Jan
195 km Cycling Time Trial        consumption-does-not-affect-the-ergogenicity-of                          Clarke, N. D.    https://pureportal.coventry.ac.uk/en/persons/neil-clarke 2021
   Histone mRNA is subject
   to 3’ uridylation and re-
   adenylation in Aspergillus https://pureportal.coventry.ac.uk/en/publications/histone-mrna-is-                           https://pureportal.coventry.ac.uk/en/persons/igor-      Feb
20nidulans                       subject-to-3-uridylation-and-re-adenylation-in-as                        Morozov, I.      morozov                                                 2021
   Improving food safety                                                                                                                                                           13
   culture in Nigeria: A review https://pureportal.coventry.ac.uk/en/publications/improving-food-                                                                                  Aug
21of practical issues            safety-culture-in-nigeria-a-review-of-practical-is                                                                                                2021
   It’s Not Just What You Do
   but the Way You Do It: A
   Systematic Review of
   Process Evaluation of
   Interventions to Improve https://pureportal.coventry.ac.uk/en/publications/its-not-just-what-you-                                                                               Dec
22Gross Motor Competence do-but-the-way-you-do-it-a-systematic-revie                                      Ma, J.           https://pureportal.coventry.ac.uk/en/persons/jiani-ma-2 2021
   Life after lockdown: The
   role of sport, exercise and
   physical activity in
   ameliorating the mental
   health implications of        https://pureportal.coventry.ac.uk/en/publications/life-after-lockdown-
23COVID-19 restrictions          the-role-of-sport-exercise-and-physical-activ                            Clarke, N. D.    https://pureportal.coventry.ac.uk/en/persons/neil-clarke 2021
   Optimal Loading for Force
   Production in the Straight
   Bar Deadlift: Force-Time                                                                                                                                                        1
   Characteristics in            https://pureportal.coventry.ac.uk/en/publications/optimal-loading-for-                    https://pureportal.coventry.ac.uk/en/persons/peter-     Jun
24Strength-Trained Adults force-production-in-the-straight-bar-deadlift                                   Mundy, P.        mundy                                                   2021
   Physical profiling of
   international cricket
   players: an investigation
   between bowlers and           https://pureportal.coventry.ac.uk/en/publications/physical-profiling-of-
25batters                        international-cricket-players-an-investigat                              Clarke, N.       https://pureportal.coventry.ac.uk/en/persons/neil-clarke 2021
   Predicting Children’s
   Energy Expenditure during
   Physical Activity using
   Deep Learning and             https://pureportal.coventry.ac.uk/en/publications/predicting-childrens-                   https://pureportal.coventry.ac.uk/en/persons/michael-
26Wearable Sensor Data           energy-expenditure-during-physical-activity-                             Duncan, M.       duncan                                                  2021
   Pre-schoolers fundamental
   movement skills predict
   BMI, physical activity, and
   sedentary behavior: A         https://pureportal.coventry.ac.uk/en/publications/pre-schoolers-                          https://pureportal.coventry.ac.uk/en/persons/michael-   Apr
   longitudinal study            fundamental-movement-skills-predict-bmi-physical-ac                                       duncan                                                  2021
27                                                                                                        Duncan, M.
   Prevalence and factors
   related to urinary
   incontinence in older
   adults women worldwide: a
   comprehensive systematic                                                                                                                                                        29
   review and meta-analysis https://pureportal.coventry.ac.uk/en/publications/prevalence-and-                              https://pureportal.coventry.ac.uk/en/persons/shadi-     Mar
28of observational studies factors-related-to-urinary-incontinence-in-older-a                             Bokaee, S.       bokaee                                                  2021
   Rates of Bile Acid
   Diarrhoea After
   Cholecystectomy: A            https://pureportal.coventry.ac.uk/en/publications/rates-of-bile-acid-                                                                             Aug
29Multicentre Audit              diarrhoea-after-cholecystectomy-a-multicentre-                                                                                                    2021
   Rationalizing teacher roles
   in developing and
   assessing physical literacy https://pureportal.coventry.ac.uk/en/publications/rationalizing-teacher-                    https://pureportal.coventry.ac.uk/en/persons/inimfon-   Feb
30in children                    roles-in-developing-and-assessing-physical-                              Essiet, I.       essiet                                                  2021
   Skin anisotropy: Finding
   the optimal incision line for
   volar forearm in males and https://pureportal.coventry.ac.uk/en/publications/skin-anisotropy-                                                                                   Dec
31females                        finding-the-optimal-incision-line-for-volar-forea                        Hill, M.         https://pureportal.coventry.ac.uk/en/persons/mathew-hill2021
   Stepping Into Salsa
   Culture: An Experiential
   Account of Engaging With                                                                                                                                                        14
   a University Non-Credit       https://pureportal.coventry.ac.uk/en/publications/stepping-into-salsa-                    https://pureportal.coventry.ac.uk/en/persons/pablo-     Sep
32Dance Programme                culture-an-experiential-account-of-engaging-w                            Domene, P.       domene                                                  2021
   Strength and Conditioning
   Practices and                                                                                                                                                                   13
   Perspectives of Volleyball https://pureportal.coventry.ac.uk/en/publications/strength-and-                              https://pureportal.coventry.ac.uk/en/persons/michael-   Feb
33Coaches and Players            conditioning-practices-and-perspectives-of-volleybal                     Duncan, M.       duncan                                                  2021
   Test-retest reliability of                                                                                                                                                       5
   soccer dribbling tests in    https://pureportal.coventry.ac.uk/en/publications/test-retest-reliability-                  https://pureportal.coventry.ac.uk/en/persons/michael-   Aug
34children                      of-soccer-dribbling-tests-in-children                                      Duncan, M.       duncan                                                  2021
   The determination of the
   lower critical concentration
   temperature and intrinsic
   viscosity: The syneresis https://pureportal.coventry.ac.uk/en/publications/the-determination-of-                         https://pureportal.coventry.ac.uk/en/persons/claire-    Mar
35reaction of polymeric gels the-lower-critical-concentration-temperature                                  Munialo, C. D.   munialo                                                 2022
   The dose–response
   relationship between
   training-load measures
   and aerobic fitness in elite https://pureportal.coventry.ac.uk/en/publications/the-doseresponse-
36academy soccer players relationship-between-training-load-measures-and-                                  Noon, M.         https://pureportal.coventry.ac.uk/en/persons/mark-noon 2021
   The effect of fatigue on
   first stance phase kinetics
   during acceleration sprint
   running in professional      https://pureportal.coventry.ac.uk/en/publications/the-effect-of-fatigue-                    https://pureportal.coventry.ac.uk/en/persons/maximilian-
37football players              on-first-stance-phase-kinetics-during-accel                                Wdowski, M.      wdowski                                                  2021
   The effects of high
   adiposity on concentric
   and eccentric muscle
   performance of upper and                                                                                                                                                         3
   lower limb musculature in https://pureportal.coventry.ac.uk/en/publications/the-effects-of-high-                         https://pureportal.coventry.ac.uk/en/persons/sharn-     Mar
38young and older adults        adiposity-on-concentric-and-eccentric-muscle-                              Shelley, S.      shelley                                                 2021
   The Impact of Student
   Attendance on
   Assessment Specific
   Performance in Sport         https://pureportal.coventry.ac.uk/en/publications/the-impact-of-                            https://pureportal.coventry.ac.uk/en/persons/steven-    Nov
39Degree Programs               student-attendance-on-assessment-specific-performan                        Eustace, S.      eustace                                                 2021
   The influence of COVID-
   19 measures in the United
   Kingdom on physical
   activity levels, perceived
   physical function and
   mood in older adults: A
   survey-based                 https://pureportal.coventry.ac.uk/en/publications/the-influence-of-                         https://pureportal.coventry.ac.uk/en/persons/michael-
40observational study           covid-19-measures-in-the-united-kingdom-on-physi                           Duncan, M. J.    duncan                                                  2021
   The Influence of
   Recruitment Age and
   Anthropometric and
   Physical Characteristics
   on the Development
   Pathway of English           https://pureportal.coventry.ac.uk/en/publications/the-influence-of-                                                                                Feb
41Academy Football Players recruitment-age-and-anthropometric-and-physical-                                Noon, M.         https://pureportal.coventry.ac.uk/en/persons/mark-noon 2021
   The influence of the urban
   food environment on diet,
   nutrition and health
   outcomes in low-income
   and middle-income                                                                                                                                                                11
   countries: A systematic                                                                                                                                                          Oct
   review                       https://pureportal.coventry.ac.uk/en/publications/the-influence-of-the-                     https://pureportal.coventry.ac.uk/en/persons/helen-     2021
42                              urban-food-environment-on-diet-nutrition-and                               Jones, H. M.     jones
   The prevalence and
   practices of caffeine use
   as an ergogenic aid in
   English professional
   soccer: Caffeine Use In      https://pureportal.coventry.ac.uk/en/publications/the-prevalence-and-
43Professional Soccer           practices-of-caffeine-use-as-an-ergogenic-aid-                             Tallis, J.       https://pureportal.coventry.ac.uk/en/persons/jason-tallis 2021
   The Relationship between
   Training Load Measures                                                                                                                                                           25
   and Next-Day Well-Being https://pureportal.coventry.ac.uk/en/publications/the-relationship-                                                                                      Jun
44in Rugby Union Players        between-training-load-measures-and-next-day-well                                                                                                    2021
   The transitioning feature
   between uncooked and
   cooked cowpea seeds
   studied by the mechanical https://pureportal.coventry.ac.uk/en/publications/the-transitioning-                           https://pureportal.coventry.ac.uk/en/persons/claire-    Mar
45compression test              feature-between-uncooked-and-cooked-cowpea-seed                            Munialo, C. D.   munialo                                                 2021
   Understanding a Player’s
   Decision-Making Process
   in Team Sports: A                                                                                                                                                                17
   Systematic Review of         https://pureportal.coventry.ac.uk/en/publications/understanding-a-                          https://pureportal.coventry.ac.uk/en/persons/mike-      May
46Empirical Evidence            players-decision-making-process-in-team-sports-a-                          Ashford, M.      ashford                                                 2021
   What Cognitive
   Mechanism, When,
   Where, and Why?
   Exploring the Decision
   Making of University and
   Professional Rugby Union                                                                                                                                                         12
   Players During               https://pureportal.coventry.ac.uk/en/publications/what-cognitive-                           https://pureportal.coventry.ac.uk/en/persons/mike-      May
47Competitive Matches           mechanism-when-where-and-why-exploring-the-decisio                         Ashford, M.      ashford                                                 2021
   A feedforward approach to
   teaching, learning and
   assessment in an
   undergraduate sports         https://pureportal.coventry.ac.uk/en/publications/a-feedforward-                                                                                   Nov
48science module                approach-to-teaching-learning-and-assessment-in-an-                        Noon, M.         https://pureportal.coventry.ac.uk/en/persons/mark-noon 2020
   Affective responses to
   supervised 10-week
   programs of resistance       https://pureportal.coventry.ac.uk/en/publications/affective-responses-                      https://pureportal.coventry.ac.uk/en/persons/michael-   Dec
49exercise in older adults      to-supervised-10-week-programs-of-resistance-                              Duncan, M.       duncan                                                  2020
   Age-related changes in
   isolated mouse skeletal
   muscle function are
   dependent on sex,                                                                                                                                                               1
   muscle, and contractility https://pureportal.coventry.ac.uk/en/publications/age-related-                                                                                        Sep
50mode                          changes-in-isolated-mouse-skeletal-muscle-function-ar                      James, R.        https://pureportal.coventry.ac.uk/en/persons/rob-james 2020
   Age-related degeneration
   of the lumbar paravertebral
   muscles: Systematic
   review and three-level        https://pureportal.coventry.ac.uk/en/publications/age-related-                         https://pureportal.coventry.ac.uk/en/persons/michael-    May
51meta-regression                degeneration-of-the-lumbar-paravertebral-muscles-syst                  Duncan, M.      duncan                                                   2020
   A national survey
   exploring the influence of
   UK-based post-graduate
   education on the past,
   present and future of                                                                                                                                                         13
   musculoskeletal               https://pureportal.coventry.ac.uk/en/publications/a-national-survey-                                                                            Nov
52physiotherapy practice.        exploring-the-influence-of-uk-based-post-gradua                                                                                                 2020
   Angle-Specific Isokinetic
   Metrics Highlight Strength                                                                                                                                                    1
   Training Needs Of Elite       https://pureportal.coventry.ac.uk/en/publications/angle-specific-                      https://pureportal.coventry.ac.uk/en/persons/steven-     Nov
53Youth Soccer Players           isokinetic-metrics-highlight-strength-training-nee                     Eustace, S.     eustace                                                  2020
   Are exercise referral
   schemes associated with
   an increase in physical
   activity? Observational
   findings using individual
   patient data meta-analysis
   from The National Referral https://pureportal.coventry.ac.uk/en/publications/are-exercise-referral-                  https://pureportal.coventry.ac.uk/en/persons/nikita-     Jun
54database                       schemes-associated-with-an-increase-in-phys                            Rowley, N.      rowley                                                   2020
   Assessing the validity and
   reliability and determining
   cut-points of the Actiwatch                                                                                                                                                   1
   2 in measuring physical https://pureportal.coventry.ac.uk/en/publications/assessing-the-                             https://pureportal.coventry.ac.uk/en/persons/laura-      Sep
55activity                       validity-and-reliability-and-determining-cut-points                    Roden, L. C.    roden                                                    2020
   Associations between self-
   reported sleep duration
   and cardiometabolic risk
   factors in young African-
   origin adults from the five-
   country Modeling the
   Epidemiologic Transition
   Study (METS)                  https://pureportal.coventry.ac.uk/en/publications/associations-                        https://pureportal.coventry.ac.uk/en/persons/laura-      Aug
56                               between-self-reported-sleep-duration-and-cardiometab                   Roden, L.       roden                                                    2020
   Calibration and Cross-
   Validation of
   Accelerometery for
   Estimating Movement                                                                                                                                                           13
   Skills in Children Aged 8- https://pureportal.coventry.ac.uk/en/publications/calibration-and-                        https://pureportal.coventry.ac.uk/en/persons/michael-    May
5712 Years                       cross-validation-of-accelerometery-for-estimating                      Duncan, M. J.   duncan                                                   2020
"""

    document2 = """
 training-specific             https://pureportal.coventry.ac.uk/en/publications/cardiorespiratory-                                                                              Aug
114physiological adaptations coordination-reveals-training-specific-physiolo                               Noon, M.         https://pureportal.coventry.ac.uk/en/persons/mark-noon 2019
   Cervical range of motion in
   school age rugby union
   players: A cross sectional https://pureportal.coventry.ac.uk/en/publications/cervical-range-of-                          https://pureportal.coventry.ac.uk/en/persons/richard-    Jul
115study                         motion-in-school-age-rugby-union-players-a-cros                           Morgan, R.       morgan-2                                                 2019
   Coffee Ingestion Improves
   5 km Cycling Performance                                                                                                                                                          25
   in Men and Women by a https://pureportal.coventry.ac.uk/en/publications/coffee-ingestion-                                                                                         Oct
116Similar Magnitude          improves-5-km-cycling-performance-in-men-and-wom                          Clarke, N.          https://pureportal.coventry.ac.uk/en/persons/neil-clarke 2019
   Critical evaluation of the
   extrapolation of data
   relative to antioxidant
   function from the
   laboratory and their
   implications on food
   production and human       https://pureportal.coventry.ac.uk/en/publications/critical-evaluation-of-                     https://pureportal.coventry.ac.uk/en/persons/claire-     May
117health: a review           the-extrapolation-of-data-relative-to-anti                                Munialo, C. D.      munialo                                                  2019
   Curcumin for cognition: is                                                                                                                                                        8
   it just hype, based on     https://pureportal.coventry.ac.uk/en/publications/curcumin-for-                                                                                        Jan
118current data?              cognition-is-it-just-hype-based-on-current-data                                                                                                        2019
   Detection of                                                                                                                                                                      1
   Heterogeneous Protein S- https://pureportal.coventry.ac.uk/en/publications/detection-of-                                 https://pureportal.coventry.ac.uk/en/persons/jennifer-   Jun
119Acylation in Cells         heterogeneous-protein-s-acylation-in-cells                                Greaves, J.         greaves                                                  2019
   Dietary Fat, Fibre,
   Satiation and Satiety – A
   Systematic Review of       https://pureportal.coventry.ac.uk/en/publications/dietary-fat-fibre-                                                                                   Mar
120Acute Studies              satiation-and-satiety-a-systematic-review-of-ac                                                                                                        2019
   Do changes in
   cardiorespiratory fitness
   resulting from physical
   activity interventions
   impact academic
   performance and
   executive function in
   children and adolescents?
   A systematic review, meta-
   analysis, and meta-        https://pureportal.coventry.ac.uk/en/publications/do-changes-in-                         https://pureportal.coventry.ac.uk/en/persons/elizabeth-
121regression                 cardiorespiratory-fitness-resulting-from-physical-a                       Horton, E.     horton                                                  2019
   Does Dietary-Induced
   Obesity in Old Age Impair
   the Contractile
   Performance of Isolated
   Mouse Soleus, Extensor
   Digitorum Longus and                                                                                                                                                       27
   Diaphragm Skeletal         https://pureportal.coventry.ac.uk/en/publications/does-dietary-                                                                                 Feb
122Muscles?                   induced-obesity-in-old-age-impair-the-contractile-pe                      James, R.      https://pureportal.coventry.ac.uk/en/persons/rob-james 2019
   Does Perception of Motor
   Competence Mediate
   Associations between
   Motor Competence and                                                                                                                                                      1
   Physical Activity in Early https://pureportal.coventry.ac.uk/en/publications/does-perception-of-                                                                          Apr
123Years Children?            motor-competence-mediate-associations-between-                            Eyre, E.       https://pureportal.coventry.ac.uk/en/persons/emma-eyre2019
   Dynamic Postural Control
   in Children: Do the Arms                                                                                                                                                     17
   Lend the Legs a Helping https://pureportal.coventry.ac.uk/en/publications/dynamic-postural-                                                                                  Jan
124Hand?                      control-in-children-do-the-arms-lend-the-legs-a-                          Hill, M.       https://pureportal.coventry.ac.uk/en/persons/mathew-hill2019
                                                                                                                                                                                13
   Editorial: Fatigue and       https://pureportal.coventry.ac.uk/en/publications/editorial-fatigue-and-                                                                        Aug
125Recovery in Football         recovery-in-football                                                     Clarke, N.    https://pureportal.coventry.ac.uk/en/persons/neil-clarke 2019
   Editorial: Cross adaptation                                                                                                                                                  8
   and cross tolerance in       https://pureportal.coventry.ac.uk/en/publications/editorial-cross-                                                                              Jan
126human health and disease adaptation-and-cross-tolerance-in-human-health-an                            Lee, B. J.    https://pureportal.coventry.ac.uk/en/persons/ben-lee     2019
   Effects of Movement
   Velocity and Training
   Frequency of Resistance
   Exercise on Functional
   Performance in Older                                                                                                                                                         7
   Adults: A Randomised         https://pureportal.coventry.ac.uk/en/publications/effects-of-movement-                 https://pureportal.coventry.ac.uk/en/persons/michael-    Feb
127Controlled Trial             velocity-and-training-frequency-of-resistance                            Duncan, M.    duncan                                                   2019
   Estimating physical activity
   in children aged 8-11
   years using accelerometry:
   contributions from
   fundamental movement
   skills and different                                                                                                                                                         18
   accelerometer                https://pureportal.coventry.ac.uk/en/publications/estimating-physical-                 https://pureportal.coventry.ac.uk/en/persons/michael-    Mar
128placements.                  activity-in-children-aged-8-11-years-using-ac                            Duncan, M.    duncan                                                   2019
   Evaluation of dietary and
   lifestyle changes as
   modifiers of S100β levels https://pureportal.coventry.ac.uk/en/publications/evaluation-of-dietary-
129in Alzheimer's disease       and-lifestyle-changes-as-modifiers-of-s100%CE%B2                                                                                                2019
   Evaporative heat loss
   insufficient to attain heat
   balance at rest in
   individuals with a spinal                                                                                                                                                   1
   cord injury at high ambient https://pureportal.coventry.ac.uk/en/publications/evaporative-heat-                                                                             Oct
130temperature                  loss-insufficient-to-attain-heat-balance-at-rest                        Price, M.      https://pureportal.coventry.ac.uk/en/persons/mike-price 2019
   Examining accelerometer
   validity for estimating
   physical activity in pre-                                                                                                                                                 1
   schoolers during free living https://pureportal.coventry.ac.uk/en/publications/examining-                                                                                 Oct
131activity                     accelerometer-validity-for-estimating-physical-activity                 Eyre, E.       https://pureportal.coventry.ac.uk/en/persons/emma-eyre2019
   Examining the Effects of
   Exercise Referral
   Schemes on Changes in https://pureportal.coventry.ac.uk/en/publications/examining-the-                              https://pureportal.coventry.ac.uk/en/persons/nikita-     Jun
132Physical Activity Levels     effects-of-exercise-referral-schemes-on-changes-in-                     Rowley, N.     rowley                                                   2019
   Examining the Presence of
   Cronobacter spp. in                                                                                                                                                          27
   Ready-to-eat Edible          https://pureportal.coventry.ac.uk/en/publications/examining-the-                       https://pureportal.coventry.ac.uk/en/persons/daniel-     Sep
133Insects                      presence-of-cronobacter-spp-in-ready-to-eat-edible-                     Amund, D.      amund                                                    2019
   Exercise intensity-
   dependent effects of arm                                                                                                                                                    21
   and leg-cycling on           https://pureportal.coventry.ac.uk/en/publications/exercise-intensity-                                                                          Oct
134cognitive performance        dependent-effects-of-arm-and-leg-cycling-on-co                          Hill, M.       https://pureportal.coventry.ac.uk/en/persons/mathew-hill2019
   Exploring the preparation
   practices of teachers who
   organise secondary school                                                                                                                                                    1
   ski trips in England and     https://pureportal.coventry.ac.uk/en/publications/exploring-the-                       https://pureportal.coventry.ac.uk/en/persons/michelle-   Feb
135Wales                        preparation-practices-of-teachers-who-organise-seco                     Stanley, M.    evans                                                    2019
   Facial Palsy Therapy: Can                                                                                                                                                    3
   Novel ‘Smart Spectacles’ https://pureportal.coventry.ac.uk/en/publications/facial-palsy-therapy-                   https://pureportal.coventry.ac.uk/en/persons/ala-         Jan
136Help People Smile?           can-novel-smart-spectacles-help-people-smile                            Szczepura, A. szczepura                                                 2019
   Higher Education                                                                                                                                                             2
   Outreach: Examining Key https://pureportal.coventry.ac.uk/en/publications/higher-education-                         https://pureportal.coventry.ac.uk/en/persons/sharon-     Oct
137Challenges For Academicsoutreach-examining-key-challenges-for-academics                              Williams, S.   williams                                                 2019
   High-Temperature
   Synthesis of Nickel-Based
   Nanoparticles for Use as
"""

    # Build Inverted-Index for documents
    inverted = {}
    documents = {'doc1':document1, 'doc2':document2}
    for doc_id, text in documents.items():
        doc_index = inverted_index(text)
        inverted_index_add(inverted, doc_id, doc_index)

    # Print Inverted-Index
    for word, doc_locations in inverted.items():
        print (word, doc_locations)

    # Search something and print results
    queries = ['challenges', 'optimal', 'help']
    for query in queries:
        result_docs = search(inverted, query)
        print ("Search for '%s': %r" % (query, result_docs))
        for _, word in word_index(query):
            def extract_text(doc, index): 
                return documents[doc][index:index+20].replace('\n', ' ')

            for doc in result_docs:
                for index in inverted[word][doc]:
                    print ('   - %s...' % extract_text(doc, index))
        print
```

```python
import re
from collections import defaultdict, Counter

def bold(txt):
    return '\x1b[1m%s\x1b[0m' % txt

DATA = [
    {
        'title': 'A Comparison of Physical Activity Between Home-Based and Centre-Based Pulmonary Rehabilitation: A Randomised Controlled Secondary Analysis',
        'Link': 'https://pureportal.coventry.ac.uk/en/publications/a-comparison-of-physical-activity-between-home-based-and-centre-b',
        'Author': 'Horton, E.',
        'Year':'26/10/2021'
    },
    {   'title': 'Actual and perceived motor competence mediate the relationship between physical fitness and technical skill performance in young soccer players',
        'Link': 'https://pureportal.coventry.ac.uk/en/publications/actual-and-perceived-motor-competence-mediate-the-relationship-be',
        'Author': 'Duncan, M.',
        'Year':'11/07/2021'
    },
]

SPLIT_RE = re.compile(r'[^a-zA-Z0-9]')
def tokenize(text):
    yield from SPLIT_RE.split(text)

def text_only(tokens):
    for t in tokens:
        if t.isalnum():
            yield t

def lowercase(tokens):
    for t in tokens:
        yield t.lower()

def stem(tokens):
    for t in tokens:
        if t.endswith('ly'):
            t = t[:-2]
        yield t

SYNONYMS = {
    'rapid': 'quick',
}
def synonyms(tokens):
    for t in tokens:
        yield SYNONYMS.get(t, t)

def analyze(text):
    tokens = tokenize(text)
    for token_filter in (text_only, lowercase, stem, synonyms):
        tokens = token_filter(tokens)
    yield from tokens

def index_docs(docs, *fields):
    index = defaultdict(lambda: defaultdict(Counter))
    for id, doc in enumerate(docs):
        for field in fields:
            for token in analyze(doc[field]):
                index[field][token][id] += 1
    return index

def combine_and(*args):
    if not args:
        return Counter()
    out = args[0].copy()
    for c in args[1:]:
        for doc_id in list(out):
            if doc_id not in c:
                del out[doc_id]
            else:
                out[doc_id] += c[doc_id]
    return out

def combine_or(*args):
    if not args:
        return Counter()
    out = args[0].copy()
    for c in args[1:]:
        out.update(c)
    return out

COMBINE = {
    'OR': combine_or,
    'AND': combine_and,
}

def search_in_fields(index, query, fields):
    for t in analyze(query):
        yield COMBINE['OR'](*(index[f][t] for f in fields))

def search(index, query, operator='AND', fields=None):
    combine = COMBINE[operator]
    return combine(*(search_in_fields(index, query, fields or index.keys())))

def query(index, query, operator='AND', fields=None):
    print('Search for "%s" using %s in %s' % (bold(query), bold(operator), fields or 'all fields'))
    print('-'*80)
    ids = search(index, query, operator, fields)
    for doc_id, score in ids.most_common():
        print('%s found with score of %s' % (bold(DATA[doc_id]['title']), bold(score)))
    print('\n')

index = index_docs(DATA, 'title', 'Link', 'Year', 'Author')

query(index, 'physical') 
query(index, '2021', fields=['Year'])
query(index, 'relationship')
```
